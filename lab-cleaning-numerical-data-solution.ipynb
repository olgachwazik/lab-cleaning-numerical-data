9. Look for outliers in the continuous variables. 
(Hint: Thereâ€™s a good plot to do that!). In case you find outliers, comment on what you will do with them.

```python
# we can see the outliers already on the plots above but in order to look at them more closely we can check boxplots:
for col in numerical_columns:
    sns.boxplot(data=customer_df, y=col)
    plt.title(col)
    plt.show()
```

We can see ouliers in customer lifetime value, premium auto and total_claim_amount columns. We might consider removing them so the model we build is not as impacted by them. Yet it might make us loose some important data (in this case about long lifetime value customers and high value customers(high monthly premium and high claim amount)). We can also use log scaling technique to scale the data and lower the impact of ouliers. 

```python
#to use the log technique: 
numericals_log = customer_df.select_dtypes(np.number)
numericals_log = numericals_log.applymap(lambda x: np.log(x+1))
```

```python
#then we can plot the numerical values to see how they changed: 
for col in numericals_log:
    sns.boxplot(data=customer_df, y=col)
    plt.title(col)
    plt.show()
```
